```python
country=['india','Germany','USA','France','Africa']
gdp=[25355,46636,34646,17377,21444]
```

```python
plt.title('country gdp using bar graph')
plt.xlabel('country')
plt.ylabel('gdp')
plt.bar(country,gdp)
plt.show()
```

```python
plt.barh(country,gdp)
plt.title('country gdp using both')
plt.xlabel('country')
plt.ylabel('gdp')
plt.show()
```
